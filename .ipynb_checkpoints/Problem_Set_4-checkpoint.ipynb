1.Positives and Average:Read 6 values that can be floating point numbers. After, print how many of them were positive. In the next line, print the average of all positive values typed, with one digit after the decimal point.

```python
count=0
temp=0
i=0
while i<=5:
    x=float(input())
    i=i+1
    if (x>=0):
        count=count+1
        temp=temp+x
print("%d valores positivos"%count)
average=float(temp/count)
print("%0.1f"%average)
    
```

3.Posative,Negative,even,odd:

```python
even=0
odd=0
pos=0
neg=0
for i in range(1,6):
    x=int(input())
    if (x%2==0):
        even=even+1
    else:
        odd=odd+1
    if (x>0):
        pos=pos+1
    elif (x<0):
        neg=neg+1
        
print("%d valor(es) par(es)"%even)  
print("%d valor(es) impar(es)"%odd)  
print("%d valor(es) positivo(s)"%pos)  
print("%d valor(es) negativo(s)"%neg)  
```

4.Odd Numbers:Read an integer value X (1 <= X <= 1000).  Then print the odd numbers from 1 to X, each one in a line, including X if is the case.

5.Six Odd Numbers:Read an integer value X and print the 6 consecutive odd numbers from X, a value per line, including X if it is the case.

```python
x = int(input())
i=0
while i<=5:
    if(x%2!=0):
        print(x)
        i=i+1
    x=x+1
```

8.Even or Odd:Read an integer value N. After, read these N values and print a message for each value saying if this value is odd, even, positive or negative. In case of zero (0), although the correct description would be "EVEN NULL", because by definition zero is even, your program must print only "NULL", without quotes.

```python
x=int(input())

for i in range(1,x+1):
    y=int(input())
    if ( y>0 and y%2==0 ):
        print("EVEN POSATIVE")
    elif ( y<0 and y%2==0 ):
        print("EVEN NEGATIVE")
    elif ( y<0 and  y%2!=0):
        print("ODD NEGATIVE")
    elif (y>0 and y%2!=0):
        print("ODD POSATIVE")
    elif( y==0):
        print("NULL")
    
```

```python
l=[]
x=int(input())

for i in range(1,x+1):
    l.append(int(input()))

for y in l:
    if ( y>0 and y%2==0 ):
        print("EVEN POSATIVE")
    elif ( y<0 and y%2==0 ):
        print("EVEN NEGATIVE")
    elif ( y<0 and  y%2!=0):
        print("ODD NEGATIVE")
    elif (y>0 and y%2!=0):
        print("ODD POSATIVE")
    elif( y==0):
        print("NULL")
```

11.Weighted Averages

```python
x=int(input())
for i in range (1,x+1):
    y=input()
    a,b,c=y.split(" ")
    a=float(a)
    b=float(b)
    c=float(c)
    temp=((a*2)+(b*3)+(c*5))/10
    print("%0.1f"%temp)
```

```python
l=[]
x=int(input())
for i in range (1,x+1):
    l.append(input())
for y in l:
    a,b,c=y.split(" ")
    a=float(a)
    b=float(b)
    c=float(c)
    temp=((a*2)+(b*3)+(c*5))/10
    print("%0.1f"%temp)
```

12.Read 10 integer numbers. Print the highest read value and the input position.
  

```python
l=[]
count=1
maxi=0
x=int(input())
for i in range (1,x+1):
    l.append(input())
for y in l:
    y=int(y)
    if (maxi<y):
        maxi=y
        count=count+1
print("Max Value:",maxi)
print("Posttion:",count)    
```

13.Maria has just started as graduate student in a medical school and she's needing your help to organize a laboratory experiment which she is responsible about. She wants to know, at the end of the year, how many animals were used in this laboratory and the percentage of each type of animal is used at all.

This laboratory uses in particular three types of animals: frogs, rats and rabbits. To obtain this information, it knows exactly the number of experiments that were performed, the type and quantity of each animal is used in each experiment.

```python
l=[]
temp1=summ=temp2=temp3=0
x=int(input())
for i in range (1,x+1):
    l.append(input())
for y in l:
    a,b=y.split(" ")
    a=int(a)
    summ =summ+a 
    if (b=='C'):
        temp1=temp1+a
    elif(b=="R"):
        temp2=temp2+a
    elif(b=="S"):
        temp3=temp3+a

print("Total:",summ,"cobaias")
print("Total de coelhos:",temp1)
print("Total de ratos:",temp2)
print("Total de sapos:",temp3)
```
